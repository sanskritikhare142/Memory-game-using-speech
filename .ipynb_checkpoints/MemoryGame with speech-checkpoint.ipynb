```python
#import section
import random, time
import speech_recognition as sr
from IPython.display import clear_output
import pyttsx3

#text to speech section
def speak(text):
    engine = pyttsx3.init()
    engine.setProperty('rate', 200)  
    voices = engine.getProperty('voices') 
    engine.setProperty('voice', voices[0].id) 
    engine.say(text)
    engine.runAndWait()
    
#obtain voice input while playing game
r = sr.Recognizer()
def recognise():
    with sr.Microphone() as source:
        print("Speak the number")
        audio = r.listen(source,timeout=3)
    try:
        sp=r.recognize_google(audio)
        if(sp=="six" or sp=="sex"):
            sp=6
        if(sp=="Tu"):
            sp=2
        s=int(sp)
        print("You said ",s)
        return s
    except Exception as e:
        print(e)
        print("Didn't get you, can you say it again?")
        return recognise()

print("Welcome to the memory game! A number will appear on your screen next and will go away after 2 seconds, you need to recall and speak the number. As you proceed to the next level, a new number will appear on the screen. You need to speak all the old numbers in order as well as the new one to clear the level. Best of luck!")
speak("Welcome to the memory game! A number will appear on your screen next and will go away after 2 seconds, you need to recall and speak the number. As you proceed to the next level, a new number will appear on the screen. You need to speak all the old numbers in order as well as the new one to clear the level. Best of luck!")    
print("Let's begin!")
speak("Lets begin!")

#obtain input from user is he/she wants to play again
def endgame():
    with sr.Microphone() as source:
        audio = r.listen(source,timeout=4)
    try:
        sp=r.recognize_google(audio)
        ans=str(sp)
        print("You said: " + ans)
        return ans
    except Exception as e:
        print(e)
        print("Didn't get you, can you say it again?")
        return endgame()

#game code
def game():
    sequence=[]
    count=0
    f=0
    score=0
    while(f==0):
        clear_output(wait=True)
        count+=1
        print("Level " +str(count))
        speak("Level " +str(count))
        clear_output(wait=True)
        #generate random number
        digit=random.randint(0,9)
        sequence.append(digit)
        print(sequence[count-1])
        time.sleep(2)
        #vanish after 5s
        clear_output(wait=True)
        #take input from user
        for i in range(0,count):
            ip=recognise()
        #check the input
            if ip!=sequence[i]:
                f=1
                print("Failed")
                break
            else:
                print("Success")
                score=score+count #Calculating the score
                
    if count==1:
        score=0
        
    print("Score: "+ str(score))
    speak("You scored"+ str(score)+"points. Congrats!")
    print("Do you want to play again? Say YES or NO")
    speak("Do you want to play again? Say YES or NO")
    answer=endgame()
    
    if(answer=="yes"):
        game()
    else:
        print("Thanks for playing! See you next time :)")
        speak("Thanks for playing! See you next time, bye!")
    
game()
```
